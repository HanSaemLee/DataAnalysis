# 라인 플롯

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# minus 표시
mpl.rcParams['axes.unicode_minus'] = False
plt.rc('font', family='Malgun Gothic')
```

```python
plt.title("라인 플롯")
plt.plot([1, 4, 9, 16])
plt.grid()
plt.show()
# x축은 자동은 0, 1, 2, 3이 된다
```

```python
plt.title("x ticks")
plt.plot([10, 20, 30, 40], [1, 4, 9 ,16]) # x축을 따로 표시 [10, 20, 30, 40]
plt.show()
```

```python
plt.title("라인 플롯")
xs = np.linspace(0, 3, 301)
ys = (xs + 1) ** 2
plt.plot(xs, ys)
plt.grid()
plt.show()
```

## 스타일 지정

```python
# 색깔(r), 마커(s), 선스타일(--)
plt.title("'rs--' 스타일의 plot")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], 'rs--')
plt.grid()
plt.show()
```

```python
# lw 선굵기, ls 선스타일, marker 마커종류, ms 마커크기, mec 마커선색깔, mew 마커선굵기, mfc 마커내부 색깔
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], c="b",
            lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.title("스타일 적용 예")
plt.grid()
plt.show()
```

## 그림 범위 지정

```python
plt.title("x축, y축의 범위 설정")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16],
            c="b", lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.xlim(0,50)   # x축의 범위
plt.ylim(-2, 20) # y축의 범위
plt.grid()
plt.show()
```

## 틱 설정

```python
# 축상의 위치 표시 지점을 틱(tick)이라 하고 
# 틱 위치를 수동으로 설정하고 싶다면 xtrics, ytricks명령을 사용한다
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("x축과 y축의 tick label 설정")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
plt.yticks([-1, 0, 1])
plt.show()
```

```python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("LaTeX, 문자열로 tick label 정의")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
            [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, 1], ["Low","Zero","High"])
plt.show()
```

## 그리드 설정

```python
# 그리드는 그래프의 특정 값을 나타내는 흐릿한 선
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("Grid 제거")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, 1], ["Low", "Zero", "High"])
plt.grid(True)
plt.show()
```

## 여러개의 선을 그리기

```python
t = np.arange(0., 5., 0.2)
plt.title("라인 플롯에서 여러개의 선 그리기")
plt.plot(t, t, 'r--', t, 0.5 * t**2, 'bs:', t, 0.2 * t**3, 'g^-') # plot(x1, y1, style1, x2, y2, style2, x3, y3, style3)
plt.show()
```

## 겹쳐그리기

```python
plt.title("라인 플롯에서 여러개의 선 그리기")
plt.plot(t, t, 'r--')
plt.plot(t, 0.5 * t**2, 'bs:')
plt.plot(t, 0.2 * t**3, 'g^-')
plt.grid()
plt.show()
```

```python
plt.title("복수의 plot 명령을 한 그림에서 표현")
plt.plot([1, 4, 9, 16],
        c="b", lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.plot([9, 16, 4, 1],
        c="k", lw=3, ls=":", marker="s", ms=10, mec="m", mew=5, mfc="c")
plt.grid()
plt.show()
```

## 범례

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.title("legend를 표시한 플롯")
plt.plot(X, C, ls="--", label="cosine")
plt.plot(X, S, ls=":", label="sine")
plt.legend(loc=2)
plt.grid()
plt.show()
```

## x축, y축 라벨, 타이틀

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.plot(X, C, label="cosine")
plt.xlabel("time")
plt.ylabel("amplitude")
plt.title("Cosine Plot")
plt.grid()
plt.show()
```

```python
# 데이터 정제하기
stopwords = ['#jeju','#일상','','#제주','#제주도','#선팔','#제주눈썹문신',
             '#서귀포눈썹문신', '#반영구', '#눈썹문신', '#제주자연눈썹',
             '#제주속눈썹', '#서귀포남자눈썹문신', '#서귀포자연눈썹', '#제주남자눈썹문신', 
             '#서귀포속눈썹', '#서귀포반영구']
'''tag_total_refined = []
for tag in tags_total:
    if tag not in stopwords:
        tag_total_refined.append(tag)'''
tags_total = [tag for tag in tags_total if tag not in stopwords]

tag_counts = Counter(tags_total)
tag_counts.most_common(50)
```

## 워드 클라우드 만들기

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# 예제ㅔ 5-22 워드클라우드 라이브러리 불러오기
import matplotlib.pyplot as plt
from wordcloud import WordCloud     # 에러시 ! pip install wordcloud 실행
import platform

if platform.system() == 'Windows':  # 윈도우의 경우
    font_path = "c:/Windows/Fonts/malgun.ttf"
elif platform.system() == 'Darwin': # Mac의 경우
    font_path = "/Users/$USER/Library/Fonts/AppleGothic.ttf"
```

```python
path = 'c:/Windows/Fonts/malgun.ttf'
wordcloud = WordCloud(font_path=path,
                      background_color='white',
                      max_words=100,
                      relative_scaling=0.3,
                      width=800, height=400
                      ).generate_from_frequencies(tag_counts)
plt.figure(figsize=(15,10))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
